```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('a.csv')

# Separate row and column labels if needed
data.set_index(data.columns[0], inplace=True)  # Set the first column as the index for row labels
numeric_data = data.apply(pd.to_numeric, errors='coerce')  # Convert data to numeric, setting non-numeric as NaN

# Set the font scale for consistent text size
sns.set(font_scale=1.2)

# Create the heatmap with three decimal places
plt.figure(figsize=(12, 6))
heatmap = sns.heatmap(numeric_data, annot=True, fmt=".3f", cbar_kws={'label': 'Score'})

# Add labels and adjust layout
plt.xlabel("Models")
plt.ylabel("Datasets")
plt.title("Model Performance Across Datasets")

# Show the heatmap
plt.savefig("graph_name.pdf",bbox_inches='tight')#, dpi=400)
plt.show()
```
