```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from random import random
def function(n):
    values = ['value1', 'value2']
    mydict = {"key " + str(i).rjust(3, '0'): int(random()*10000) for i in range(n)}

    return mydict
```

```python
def make_graph(hist,step):
    title="____title____"
    x=[str(x) for x in list(hist.keys())] 
    if len(hist)>30:
        plt.figure(figsize=(28,8))
        plt.xticks(range(1,len(x),step),rotation=90)
    plt.bar(x, hist.values(), align='center',color='#607c8e')
    plt.title(title)
    plt.xticks(rotation=90) 
    plt.xlabel('Packet Size')
    plt.ylabel('Counts')
    plt.grid(axis='y', alpha=0.75)
    plt.grid(axis='x', alpha=0.75)
    graph_name=f"{title}.pdf"
    plt.savefig(graph_name,bbox_inches='tight',format="pdf")#, dpi=400)
    plt.show()
```

```python
hist=function(30)
make_graph(hist,1)
```

```python
hist=function(300)
make_graph(hist,1)
```

```python
hist=function(300)
make_graph(hist,10)
```
